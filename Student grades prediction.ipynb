# Student grades prediction

Project Description
The dataset contains grades scored by students throughout their university tenure in various courses and their CGPA calculated based on their grades
Columns Description-  total 43 columns
-Seat No : The enrolled number of candidate that took the exams

CGPA : The cumulative GPA based on the four year total grade progress of each candidate . CGPA is a Final Marks -- provided to student.
 
· All other columns are course codes in the format AB-XXX where AB are alphabets representing candidates' departments and XXX are numbers where first X represents the year the canditate took exam

 
Predict - CGPA of a student based on different grades in four years.

Dataset Link-
•  https://github.com/FlipRoboTechnologies/ML-Datasets/blob/main/Grades/Grades.csv

```python
import pandas as pd

# Load the dataset
file_path = "https://raw.githubusercontent.com/FlipRoboTechnologies/ML-Datasets/main/Grades/Grades.csv"
df = pd.read_csv(file_path)
```

```python
print(df['CGPA'].value_counts(bins=10, sort=False))
```

```python
# Explore the relationship between different course grades and CGPA
# For example, let's look at the correlation between the first course grade and CGPA
first_course_grade = df.columns[2]  # Assuming 'Seat No' and 'CGPA' are the first two columns
print(df[[first_course_grade, 'CGPA']].corr())
```

```python
# You can also visualize the data using matplotlib or seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Scatter plot of the first course grade vs CGPA
sns.scatterplot(x=first_course_grade, y='CGPA', data=df)
plt.title('First Course Grade vs CGPA')
plt.show()
```

```python
# Heatmap of the correlation matrix for all numerical columns
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
```

```python
# Step 2: Handle missing values (if any)
# For visualization, we might want to know if there are missing values, so we'll check first
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values[missing_values > 0])
```

```python
# If there are missing values, we can choose to drop rows with missing values or fill them
# For example, let's fill missing numerical values with the median of the column
df = df.fillna(df.median())
```

```python
# Step 3: Convert grades to a numeric scale if they are letter grades
# This step is necessary if the grades are not already in a numeric format
if 'Grade' in df.columns:
    grade_mapping = {'A': 4, 'B': 3, 'C': 2, 'D': 1, 'F': 0}
    df['Grade'] = df['Grade'].map(grade_mapping)
```

```python
# Example visualization: Distribution of CGPA
sns.displot(df['CGPA'], kde=True)
plt.title('Distribution of CGPA')
plt.show()
```
